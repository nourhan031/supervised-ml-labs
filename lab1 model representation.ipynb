Problem Statement:
housing price prediction.
This lab will use a simple data set with only two data points - a house with 1000 square feet(sqft) sold for $300,000 and a house with 2000 square feet sold for $500,000. These two points will constitute our data or training set. In this lab, the units of size are 1000 sqft and the units of price are 1000s of dollars.

Size (1000 sqft) 	Price (1000s of dollars)
1.0 	             300
2.0 	             500

fit a linear regression model (shown above as the blue straight line) through these two points, so you can then predict price for other houses - say, a house with 1200 sqft.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0,2.0]) # input feature
y_train = np.array([300.0,500.0]) # target

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# m: number of training examples
# .shape: returns a python tuple with an entry for each 
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]  # length of array and number of examples
print(f"number of training examples: {m}")
```

```python
# or len()
m = len(x_train)
print(f"number of training examples: {m}")
```

```python
# training example
# (x_i,y_i): ith training example
# (x(0),y(0))=(1,300)
# (x(1),y(1))=(2,500)
for i in range(len(x_train)):
    x_i = x_train[i]
    y_i = y_train[i]
    print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
# plotting the data
# args 'marker' and 'c': show the pts as red cross (the default is blue dots)

# plotting data pts
plt.scatter(x_train,y_train,marker='x',c='r')
plt.title("Housing Prices")
plt.ylabel('Price (in $1000s)')
plt.xlabel('Size (1000 sqft)')
plt.show()
```

MODEL FUNCTION
洧녭洧녻,洧녪(洧논(洧녰))=洧녻洧논(洧녰)+洧녪

```python
# giving w and b initial values to see what happens
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

for 洧논(0), f_wb = w * x[0] + b
for 洧논(1), f_wb = w * x[1] + b

therefore:
for x(i) -> f_wb = w * x[i] + b

The argument description (ndarray (m,)) describes a Numpy n-dimensional array of shape (m,). 

(scalar) describes an argument without dimensions, just a magnitude.

np.zero(n) will return a one-dimensional numpy array with 洧녵 entries

```python
"""
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
    """
```

```python
def compute_output(x,w,b):
    m = x.shape[0]
    f_wb = np.zeros(m)

    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb        
```

```python
tmp_f_wb = compute_output(x_train,w,b)

# plot model prediction
plt.plot(x_train,tmp_f_wb,c='b',label='my prediction')

# plot data pts
plt.scatter(x_train,y_train,marker='x',c='r',label='actual values')

plt.title("Housing Prices")
plt.ylabel('Price (in $1000s)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
